## Exercici 11

Crea una nova columna anomenada Person a on els valors siguin (male,female,child). El child haurien de ser aquells passatgers menors de 16 anys.

```python
#passatgers[passatgers.Person == 'Child']
def get_age_group(row):
    threshold = 16
    if row.Age <= int(threshold):
        row.Person = 'child'
    else:
        row.Person = row.Sex
    return row

passatgers.apply(get_age_group, axis='columns')
passatgers
```
