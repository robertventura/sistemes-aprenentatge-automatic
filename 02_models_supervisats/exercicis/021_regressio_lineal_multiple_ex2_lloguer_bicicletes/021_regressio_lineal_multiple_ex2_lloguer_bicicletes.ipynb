## Introducció

Volem utilitzar la tècnica de Regressió lineal múltiple per predir el número de lloguers de bicicletes segons una sèrie d'atributs.
El dataset és [daily-bike-share.csv](daily-bike-share.csv)

Aquest dataset ha esat extret de [Bike Sharing Dataset de UCI University Californi Irvine] (https://archive.ics.uci.edu/ml/datasets/bike+sharing+dataset)

* **instant:** Preu de la vivenda en dollars
* **dteday:** data (en format americà (mes/dia/any)
* **season:** Estació de l'any (1:hivern,2:primavera,3:estiu,4:tardor)
* **yr:** Any (0:2011, 1:2012)
* **mnth:** mes (1-12)
* **holiday:** Indica si el dia és festiu o no
* **weekday:** Indica el dia de la setmana (diumenge=0,dilluns=1,...dissabte=6)
* **workingday:** Si el dia és laborable (dilluns-divendres) i tampoc és fesiu 
* **wheathersit:** Temps 
    * **1:** Clar, Pocs núvols, Parcialment ennuvolat, Parcialment ennuvolat
    * **2:** Boira + Ennuvolat, Boira + Núvols trencats, Boira + Pocs núvols, Boira
    * **3:** Neu feble, pluja lleugera + tempesta + núvols dispersos, pluja lleugera + núvols dispersos
    * **4:** Pluja intensa + granisada + tempesta elèctrica + boira, neu + boira
* **temp:** Temperatura normalitzada en graus centígrads.
* **atemp:** Temperatura de sensació /confort normalitzada en graus centígrads
* **hum:** Humitat normalitzada en tant per 1
* **windspeed:** Velocitat del vent normalitzada.(màx 63). Creiem que en milles/h (mph). 1 kph = 0.621371192 mph
* **rentals:** Quantitat de lloguers
